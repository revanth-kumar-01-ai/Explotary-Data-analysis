# Iris Project 

Iris Dataset is a popular dataset among beginners in Data Science. It contains data of about 150 Iris Flowers and their attributes which include 
- sepal length
-  sepal width
-  petal length
-  petal width.

## Problem Description 
This project, the goal is to analyze the correlation between these attributes and the three types of Iris Flowers: Iris-Setosa, Iris-Versicolour, and Iris-Virginica.

# Load the Iris dataset from sklearn.datasets

```python
#  import the libraries 

import pandas as pd 
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt 
import seaborn as sns
```

# Convert Pandas Dataframe 

```python
# load the data 
irisDataset = load_iris()
```

```python
df = pd.DataFrame(data=irisDataset.data, columns=irisDataset.feature_names)
```

```python
df.head()
```

```python
df.describe()
```

# Visualize the correlation matrix using a heatmap.

```python
# Add the traget column 

df['species'] = irisDataset.target 


df['species'] = df['species'].map({
                0:'Iris-Setosa', 
                1:'Iris-Versicolour', 
                3:'Iris-Virginica'
})
```

```python
df.head()
```

```python
correlationMatrix = df.iloc[:, :-1].corr()
```

```python
correlationMatrix
```

```python
# plot heatmap 

sns.heatmap(correlationMatrix, annot = True, cmap = 'coolwarm', linewidths=0.9)
plt.title("Correlation Matrix of IRIS dataset")
plt.show()
```

```python
# Pairplot to visualize attribute relationships 
sns.pairplot(df, hue="species", diag_kind="kde", palette="Set2")
plt.show()
```
